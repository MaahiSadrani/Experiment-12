```python
# Experiment 12:Categorial Data Analysis using Python:

# Second dataset
import pandas as pd

df = pd.read_csv("/content/Expt11.csv")
print(df)
```

```python
# 1. To count the number of students in Each grade
# Find how many students obtained each grade

print(df['Grade'].value_counts())
```

```python
# 2. Number of students in each Gender
print(df['Gender'].value_counts())
```

```python
# 3. Number of students in each department
print(df['Department'].value_counts())
```

```python
# 4. To find the percentage Distribution of grades:

print(df['Grade'].value_counts(normalize=True)*100)
```

```python
# 5. Grade vs Gender analysis:

print(pd.crosstab(df['Gender'],df['Grade']))
```

```python
#6. Department vs gender:

print(pd.crosstab(df['Department'],df['Gender']))
```

```python
#7. Department vs Grade Percentage

print(pd.crosstab(df['Department'],df['Grade'],normalize='index')*100)
```

```python
#8. grouping data
print(df.groupby('Department')['Grade'].value_counts())
```

```python
# Cross Tabulation

data = {
'Order_ID':['O1','O2','O3','O4','O5','O6','O7','O8','O9','O10'],

'Category':['Electronics','Clothing','Electronics','Grocery','Clothing',
            'Electronics','Grocery','Clothing','Electronics','Grocery'],

'Payment_Method':['UPI','Card','UPI','Cash','Card',
                  'UPI','Cash','Card','UPI','Cash'],

'Delivery_Type':['Express','Standard','Express','Standard','Express',
                 'Standard','Express','Standard','Express','Standard'],

'Customer_Type':['New','Returning','Returning','New','Returning',
                 'New','Returning','New','Returning','New']
}
df = pd.DataFrame(data)
print(df)
```

```python
#1. Frequency Count
# Count how many orders in each category

print(df['Category'].value_counts())
```

```python
#2. Frequency of payment method

print(df['Payment_Method'].value_counts())
```

```python
#3. Percentage distribution

print(df['Category'].value_counts(normalize=True)*100)
```

```python
#4. Filtering Data
# Example: Order from Electronics category

electronics_orders = df[df['Category'] == 'Electronics']
print(electronics_orders)
```

```python
#5. Sorting data

sorted_data = df.sort_values(by='Category')
print(sorted_data)
```

```python
#6. Unique value
# display all unique values

print(df['Category'].unique())
```

```python
#7. Number of Unique values

print(df['Category'].nunique())
```

```python
#8. Cross Tabulation
# Category vs Payment method
print(pd.crosstab(df['Category'],df['Payment_Method']))
```

```python
#9. Grouping
# Ordsers by category and payment method

print(df.groupby('Category')['Payment_Method'].value_counts())
```
